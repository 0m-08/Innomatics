```python
def analyze_post_engagement(likes_list):
    """
    Function to calculate total likes and determine post status.
    Developed by Dongare
    """
    
    total_likes = 0
    
    # Using loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Conditional logic to determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    return total_likes, status


# Sample real-time data (likes from different platforms or time intervals)
likes_data = [250, 300, 150, 200, 150]

# Function call
total, post_status = analyze_post_engagement(likes_data)

# Output
print("Developed by: Dongare")
print("Total Likes:", total)
print("Post Status:", post_status)
```

```python
def analyze_rainfall(rainfall_data, required_level):
    """
    Function to calculate average rainfall
    and determine adequacy status.
    Developed by Dongare
    """
    
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain
    
    # Mathematical operation to calculate average
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Conditional logic to determine rainfall status
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    return average_rainfall, status


# Example seasonal rainfall data (in mm)
season_data = [70, 75, 68, 80, 67]

# Required rainfall level
required_rainfall = 70

# Function call
average, rainfall_status = analyze_rainfall(season_data, required_rainfall)

# Output
print("Developed by: Dongare")
print("Average Rainfall:", round(average))
print("Rainfall Status:", rainfall_status)
```

```python
def detect_duplicate_accounts(usernames):
    """
    Function to detect duplicate usernames.
    Developed by Dongare
    """
    
    unique_usernames = set()
    
    # Loop through usernames to check duplicates
    for name in usernames:
        if name in unique_usernames:
            return "Yes"
        unique_usernames.add(name)
    
    return "No"


# Example username data
user_list = ["rahul_01", "anita_22", "rahul_01", "vikas_77"]

# Function call
duplicate_status = detect_duplicate_accounts(user_list)

# Output
print("Developed by: Dongare")
print("Duplicate Accounts Found:", duplicate_status)
```

```python
def filter_premium_crops(price_list, threshold):
    """
    Function to filter crop prices above premium threshold.
    Developed by Dongare
    """
    
    premium_crops = []
    
    # Loop to filter premium crop prices
    for price in price_list:
        if price > threshold:
            premium_crops.append(price)
    
    return premium_crops


# Example crop prices (in ₹)
crop_prices = [1800, 2500, 1500, 3200, 2000]

# Premium threshold
premium_threshold = 2000

# Function call
premium_list = filter_premium_crops(crop_prices, premium_threshold)

# Output
print("Developed by: Dongare")
print("Premium Crops:", premium_list)
```

```python
def check_application_health(error_count):
    """
    Function to evaluate application health based on error count.
    Developed by Dongare
    """
    
    # Conditional statements to determine system status
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    return error_count, status


# Example error count
current_errors = 7

# Function call
errors, system_status = check_application_health(current_errors)

# Output
print("Developed by: Dongare")
print("Error Count:", errors)
print("System Status:", system_status)
```

```python
def check_transaction_limit(amount):
    """
    Function to validate daily transaction amount.
    Developed by Dongare
    """
    
    daily_limit = 50000  # ₹50,000 limit
    
    # Conditional logic to validate transaction
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    return amount, status


# Example transaction amount
transaction_amount = 60000

# Function call
amount, transaction_status = check_transaction_limit(transaction_amount)

# Output
print("Developed by: Dongare")
print("Transaction Amount:", amount)
print("Transaction Status:", transaction_status)
```

```python
def check_attendance_eligibility(attendance_list):
    """
    Function to calculate attendance percentage and determine exam eligibility.
    Developed by Dongare
    """
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Loop to count attended classes
    for status in attendance_list:
        if status == "P":  # P = Present
            attended_classes += 1
    
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    
    # Conditional logic for exam eligibility
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    
    return round(attendance_percentage, 1), eligibility


# Example attendance data ('P' for present, 'A' for absent)
student_attendance = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]

# Function call
percentage, exam_status = check_attendance_eligibility(student_attendance)

# Output
print("Developed by: Dongare")
print("Attendance Percentage:", percentage)
print("Exam Eligibility:", exam_status)
```

```python
def calculate_electricity_bill(units_consumed):
    """
    Function to calculate electricity bill and classify usage level.
    Developed by Dongare
    """
    
    # Initialize bill
    bill = 0
    
    # Billing slabs
    if units_consumed <= 100:
        bill = units_consumed * 3
    elif units_consumed <= 200:
        bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status


# Example units consumed
units = 350

# Function call
total_bill, usage_level = calculate_electricity_bill(units)

# Output
print("Developed by: Dongare")
print("Total Bill: ₹", total_bill)
print("Usage Classification:", usage_level)
```
