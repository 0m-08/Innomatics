## 1. Introduction to LangChain

### 1.1 What is LangChain?

LangChain is an open-source framework that helps developers build applications powered by large language models (LLMs). It was created by Harrison Chase in October 2022 and quickly became one of the most popular AI frameworks out there. The core idea behind LangChain is pretty simple — instead of just making a single call to a language model, you want to chain multiple components together to create something actually useful.

At its heart, LangChain provides a set of modular components that you can compose together: LLMs, prompts, chains, memory, agents, tools and document loaders. Think of it like LEGO blocks — each piece does one specific thing and you connect them to build a bigger structure.

**Key Definition:** LangChain = LLM + Orchestration + Memory + Tools + Agents, all in one framework

### 1.2 Why is LangChain Important in the Modern LLM Ecosystem?

When GPT-3 and GPT-4 came out, everyone got excited about what LLMs could do. But using them in real applications was actually quite painful. You had to manually handle prompt engineering, manage conversation history, figure out how to connect to external databases, and write a ton of boilerplate code just to do basic stuff.

LangChain basically solved all these problems in one go. It became the de-facto standard for building LLM-powered applications because:
- It abstracts away low-level API calls and gives you clean, reusable components
- It supports almost every major LLM provider — OpenAI, Anthropic, HuggingFace, Cohere, etc.
- It has a huge ecosystem of integrations — over 100+ tools, databases, and APIs
- The community is massive, so you'll find solutions to most problems easily
- It keeps up with the fast-moving AI space with regular updates

### 1.3 Problems LangChain Solves

| Problem | Description |
|---------|-------------|
| ORCHESTRATION | Managing the flow between multiple LLM calls, tools, and data sources |
| CHAINING | Connecting outputs of one component as inputs to the next automatically |
| MEMORY | Keeping track of conversation history across multiple interactions |
| TOOL INTEGRATION | Easily connecting LLMs to search engines, databases, APIs |
| PROMPT MANAGEMENT | Templating and versioning prompts in a structured way |
| AGENT BEHAVIOR | Letting the LLM decide which tools to use and when |

## 2. Core Components of LangChain

LangChain is build around a set of well-defined components. Each one handles a specific concern and they all work together seamlessly. Let's go through each one with code examples.

2.1 LLMs and Chat Models
LLMs & Chat Models	LLMs (Language Models) take a plain string as input and return a string. Chat Models are more structured — they take a list of messages (System, Human, AI) and return a message. Most modern models like GPT-4 and Claude use the Chat Model interface.
Why it exists: So we don't have to write different integration code for each provider. LangChain gives a unified interface regardless of which model backend you're using.


````python
# Install: pip install langchain langchain-openai

from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

# Initialize Chat Model
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

# Create messages
messages = [
    SystemMessage(content="You are a helpful AI assistant."),
    HumanMessage(content="What is LangChain in one sentence?")
]

# Invoke the model
response = llm.invoke(messages)
print(response.content)

# Output: LangChain is a framework for building LLM-powered apps.
````

### 2.2 Prompts and Prompt Templates
Prompts & Templates	A PromptTemplate lets you define a reusable prompt with placeholder variables. Instead of hardcoding prompts everywhere in your code, you create a template once and fill in the variables at runtime. ChatPromptTemplate works the same way but for chat-based models.
Why it exists: Hardcoded prompts are a maintenance nightmare. Templates separate the prompt logic from application logic and make it much easier to iterate on prompt design.



````python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

# Basic PromptTemplate
template = PromptTemplate(
    input_variables=["topic", "level"],
    template="Explain {topic} to a {level} student in simple terms."
)

# Format the prompt
prompt = template.format(topic="neural networks", level="high school")
print(prompt)
# -> Explain neural networks to a high school student in simple terms.

# ChatPromptTemplate (modern approach)
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in {domain}."),
    ("human", "Explain {concept} in simple terms.")
])

messages = chat_template.format_messages(
    domain="machine learning",
    concept="gradient descent"
)
print(messages)
````

### 2.3 Chains
Chains	A Chain is what you get when you connect components together using the pipe operator (|) or the LangChain Expression Language (LCEL). The output of one component flows into the next automatically. This is the core pattern of LangChain — building pipelines of components.
Why it exists: Without chains, you'd manually pass outputs from one function to another. Chains make this declarative and composable, and they support streaming, batching, and async out of the box.


````python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Define components
llm = ChatOpenAI(model='gpt-4o-mini')
parser = StrOutputParser()

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a Python expert."),
    ("human", "Write a short function to {task}")
])

# Chain them using the pipe operator (LCEL)
chain = prompt | llm | parser
# Invoke the chain
result = chain.invoke({"task": "reverse a string"})
print(result)

# Batch processing (multiple inputs at once)
results = chain.batch([
    {"task": "sort a list"},
    {"task": "find max in a list"}
])
````

### 2.4 Memory
Memory	Memory components store the history of a conversation and inject it into subsequent prompts. Without memory, every LLM call is stateless — the model forgets everything said before. LangChain provides different memory types: ConversationBufferMemory, ConversationSummaryMemory, etc.
Why it exists: Real-world chatbots and assistants need to remember context. Memory abstracts the complexity of managing and injecting conversation history into prompts.


### 2.8 Indexes and Vector Stores

**Vector Stores:** A Vector Store stores document embeddings (numerical representations of text) so you can do semantic similarity search. When a user asks a question, their query is converted to an embedding and compared to all stored embeddings to find the most relevant documents. FAISS and Chroma are commonly used.

**Why it exists:** LLMs have limited context windows — you can't dump an entire 500-page PDF into a prompt. Vector stores let you retrieve only the most relevant chunks of information for a given query.

````python
# 2.7 Example: Document Loaders
from langchain_community.document_loaders import PyPDFLoader, WebBaseLoader
from langchain_community.document_loaders.csv_loader import CSVLoader

# Load from PDF
pdf_loader = PyPDFLoader("sample_document.pdf")
pdf_docs = pdf_loader.load()
print(f'Loaded {len(pdf_docs)} pages from PDF')
print(pdf_docs[0].page_content[:200])  # First 200 chars

# Load from a website
web_loader = WebBaseLoader('https://python.langchain.com/docs/introduction')
web_docs = web_loader.load()
print(f'Loaded {len(web_docs)} documents from web')

# Load from CSV
csv_loader = CSVLoader("data.csv")
csv_docs = csv_loader.load()

# Each Document has .page_content and .metadata
for doc in csv_docs[:2]:
    print(doc.page_content)
    print(doc.metadata)  # -> {'source': 'data.csv', 'row': 0}
````

### 2.7 Document Loaders

**Document Loaders:** Document Loaders are responsible for loading data from various sources into a standardized Document format. LangChain has loaders for PDFs, CSVs, web pages, Notion, Google Drive, GitHub, YouTube transcripts, and many more.

**Why it exists:** Before you can do anything useful with external data (like Q&A over documents), you need to load it. Loaders provide a consistent interface regardless of the source format.

````python
# 2.6 Example: Built-in Tools
from langchain_community.tools import DuckDuckGoSearchRun, WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain_core.tools import tool

# Built-in web search tool
search = DuckDuckGoSearchRun()
print(search.run("LangChain latest version 2024"))

# Wikipedia tool
wiki = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
print(wiki.run("LangChain framework"))

# Custom tool with @tool decorator
@tool
def get_word_count(text: str) -> int:
    """Count the number of words in a given text string."""
    return len(text.split())

# Tools have metadata
print(get_word_count.name)         # -> get_word_count
print(get_word_count.description)  # -> Count the number of words...
print(get_word_count.args)         # -> {'text': {'type': 'string'}}
````

### 2.6 Tools

**Tools:** Tools are functions that agents can call to interact with the outside world. LangChain has built-in tools like DuckDuckGoSearch, Wikipedia, Python REPL, and Calculators. You can also create custom tools using the @tool decorator.

**Why it exists:** LLMs are trained on static data — they can't search the web or run code. Tools bridge this gap by giving the LLM access to real-time external capabilities.

````python
# 2.5 Example: ReAct Agent with Tools
from langchain_openai import ChatOpenAI
from langchain.agents import create_react_agent, AgentExecutor
from langchain_core.tools import tool
from langchain import hub

# Define custom tools
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two integers together."""
    return a * b

@tool
def add(a: int, b: int) -> int:
    """Add two integers together."""
    return a + b

tools = [multiply, add]

# Load a ReAct prompt from LangChain Hub
prompt = hub.pull("hwchase17/react")

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)

# Create the agent
agent = create_react_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Run the agent
result = agent_executor.invoke({
    "input": "What is (15 * 4) + 20?"
})
print(result['output'])  # -> 80
````

````python
# 4.5 Memory Example — Multi-turn Conversation
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model='gpt-4o-mini')

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a friendly tutor. Help the student learn step by step."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | llm | StrOutputParser()

# Create session store
sessions = {}

def get_history(session_id):
    if session_id not in sessions:
        sessions[session_id] = InMemoryChatMessageHistory()
    return sessions[session_id]

chat = RunnableWithMessageHistory(
    chain, get_history,
    input_messages_key="input",
    history_messages_key="chat_history"
)

cfg = {'configurable': {'session_id': 'student1'}}

# Multi-turn conversation
print(chat.invoke({'input': 'I want to learn Python. Where do I start?'}, config=cfg))
print(chat.invoke({'input': 'How long will that take?'}, config=cfg))
print(chat.invoke({'input': 'What did I say I wanted to learn?'}, config=cfg))
# -> The model correctly answers 'Python' because it remembers!
````

````python
# 4.4 Agent with Tool
from langchain_openai import ChatOpenAI
from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.tools import tool
import datetime

@tool
def get_current_date() -> str:
    """Returns todays current date and time."""
    return datetime.datetime.now().strftime('%Y-%m-%d %H:%M')

@tool
def celsius_to_fahrenheit(celsius: float) -> float:
    """Convert a temperature from Celsius to Fahrenheit."""
    return round((celsius * 9/5) + 32, 2)

@tool
def simple_calculator(expression: str) -> str:
    """Evaluate a simple math expression like '2+2' or '10*5'."""
    try:
        return str(eval(expression))
    except:
        return 'Invalid expression'

tools = [get_current_date, celsius_to_fahrenheit, simple_calculator]

# Prompt with message history placeholder
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Use tools when needed."),
    ("human", "{input}"),
    MessagesPlaceholder("agent_scratchpad")
])

llm = ChatOpenAI(model='gpt-4o-mini', temperature=0)
agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Ask something that needs tools
result = executor.invoke({'input': 'What is today date and what is 37 celsius in fahrenheit?'})
print(result['output'])
````

````python
# 4.3 Simple Chain — Q&A with RAG
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_core.documents import Document

# Sample knowledge base
docs = [
    Document(page_content="LangChain was created by Harrison Chase in Oct 2022."),
    Document(page_content="LCEL stands for LangChain Expression Language."),
    Document(page_content="ReAct agents use Reason + Act loops to solve tasks."),
    Document(page_content="FAISS is Facebook AI Similarity Search library."),
]

# Build vector store
embeddings = OpenAIEmbeddings()
vectorstore = FAISS.from_documents(docs, embeddings)
retriever = vectorstore.as_retriever(search_kwargs={'k': 2})

# RAG Prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer using only this context:\n\n{context}"),
    ("human", "{question}")
])

llm = ChatOpenAI(model='gpt-4o-mini')

# Build the RAG chain
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt | llm | StrOutputParser()
)

# Test it
answer = rag_chain.invoke("Who created LangChain?")
print(answer)  # -> LangChain was created by Harrison Chase in October 2022.
````

## 3. Architecture Explanation

Now that we understand all the individual components, let's look at how they all fit together. The overall architecture of a LangChain application follows a clean pipeline pattern where each component handles one specific responsibility.

### 3.1 The LangChain Pipeline Flow

```
👤 USER INPUT
    ↓
📝 PROMPT TEMPLATE (Input is injected into a structured prompt template with variables)
    ↓
🧠 LLM / CHAT MODEL (The formatted prompt is sent to the LLM - GPT-4, Claude, etc.)
    ↓
⛓️ CHAIN (LCEL) (Output passes through the pipeline, potentially to multiple steps)
    ↓
🔧 TOOL / AGENT (If needed, agent decides to call external tools - search, calc, API)
    ↓
🧩 MEMORY / VECTOR STORE (Results combined with memory context and retrieved documents)
    ↓
✅ FINAL OUTPUT (Parsed, structured response delivered back to the user)
```

### 3.2 RAG (Retrieval-Augmented Generation) Architecture

A very common architecture pattern in LangChain is RAG. Here the document retrieval step is added before the LLM call:

```
Documents → Load → Split into Chunks → Embed → Store in Vector DB
User Query → Embed Query → Similarity Search → Retrieve Top-K Chunks
Retrieved Chunks + Query → Prompt Template → LLM → Answer
```

This allows the LLM to answer questions based on YOUR specific documents, not just its training data. Very useful for internal knowledge bases.
